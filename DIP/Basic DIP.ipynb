 # Image Mask 

```python
img_1 = plt.imread('ImageSet/4.1.01.tiff')
nrows, ncols, channels = img_1.shape
row, col = np.ogrid[:nrows,:ncols]
center_row, center_col = nrows/2, ncols/2

outer_disk_mask = ((row - center_row)**2 +(col-center_col)**2 > (nrows/2)**2)
img_1.setflags(write=1)
img_1[outer_disk_mask] = 0

plt.imshow(img_1)
plt.axis('off')
plt.show()
```

# Image Channels

```python
img3 = plt.imread('ImageSet/4.1.04.tiff')
r = img3[:,:,0]
g = img3[:,:,1]
b = img3[:,:,2]

output = [img3, r, g, b]
titles = ['Image', 'Red', 'Green','Blue']
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.axis('off')
    plt.title(titles[i])
    if i==0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i],cmap='gray')
        
plt.show()
```

# Arithmetic Ops

```python
img1 = plt.imread('ImageSet/4.1.06.tiff')
img2 = plt.imread('ImageSet/4.1.04.tiff')
plt.imshow(img1)
plt.show()
plt.imshow(img2)
plt.show()
```

```python
plt.imshow(img1+img2)
plt.show()

plt.imshow(img2+img1)
plt.show()

#Addition is commutative, Hence outputs will be same
```

```python
plt.imshow(img1-img2)
plt.show()

plt.imshow(img2-img1)
plt.show()

#Subtraction is not commutative.
```

```python
plt.imshow(np.flip(img1,-1))
plt.show()

plt.imshow(np.flip(img1,-2))
plt.show()

plt.imshow(np.flip(img1,0))
plt.show()

plt.imshow(np.flip(img1,1))
plt.show()

plt.imshow(np.flip(img1,2))
plt.show()
```

```python
plt.imshow(np.roll(img1,256))
plt.show()

plt.imshow(np.roll(img1,2048))
plt.show()
```

```python
plt.imshow(np.fliplr(img1))
plt.show()

plt.imshow(np.flipud(img1))
plt.show()

plt.imshow(np.rot90(img1))
plt.show()
```

# Logical Ops

```python
img1 = plt.imread('ImageSet/4.1.06.tiff')
img2 = plt.imread('ImageSet/4.1.04.tiff')

plt.imshow(np.bitwise_and(img1,img2)) #Commutative property
plt.show()

plt.imshow(np.bitwise_or(img1,img2)) #Commutative property
plt.show()

#plt.subplot(1,2,1)
plt.imshow(np.bitwise_xor(img1,img2)) #Commutative property
plt.show()

plt.imshow(np.bitwise_not(img1))
plt.show()
plt.imshow(np.bitwise_not(img2))
plt.show()
```

# Image Histograms with Numpy and Matplotlib

```python
img1 = plt.imread('ImageSet/4.1.01.tiff')

r = img1[:,:,0]
g = img1[:,:,1]
b = img1[:,:,2]
```

```python
plt.subplots_adjust(hspace=0.5,wspace=0.5)

plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(img1)

hist, bins = np.histogram(r.ravel(),bins=256,range=(0,256))

plt.subplot(2,2,2)
plt.title('Red Histogram')
plt.bar(bins[:-1],hist)
plt.show()

hist, bins = np.histogram(g.ravel(),bins=256,range=(0,256))
plt.subplot(2,2,3)
plt.title('Green Histogram')
plt.bar(bins[:-1],hist)
plt.show()

hist, bins = np.histogram(b.ravel(),bins=256,range=(0,256))
plt.subplot(2,2,4)
plt.title('Blue Histogram')
plt.bar(bins[:-1],hist)
plt.show()
```
